```python
# Create a sine wave of 100 sample points
n_points = 100
```

```python
n_points
```

```python
import numpy as np
x = np.linspace(0, 1, num = n_points)
```

```python
x
```

```python
# Plot the data points using sine wave
z = np.sin(2*np.pi*x)
```

```python
import matplotlib.pyplot as plot
plot.figure(figsize = (10, 8))
```

```python
# g represents the color of the plot -- Green
plot.plot(x, z, 'g-') 
```

```python
# Create a Gaussian noise
eta = .4*np.random.normal(0, 1, n_points) # Add gaussian noise of mean 0 and standard deviation 1
# Plot the noise
plot.figure
plot.plot(x, eta, 'g-')
```

```python
# Add noise to the since wave to create noisy sine wave
t = z + eta

# Plot
plot.figure
plot.plot(x, z, 'g-', x, t, 'bo')
plot.legend(('Sine Wave', 'Noisy Signal'))
```

```python
# Sample 10 points from the noise sine wave
import numpy as np
sample_points = 10

x = np.linspace(0, 1, num = sample_points)
eta = 0.2*np.random.normal(0, 1, sample_points) # Add noise of mean 0 and standard deviation 1
# t <-- Represents noisy sine wave of 10 sampled points
t = np.sin(2*np.pi*x) + eta

xf = np.linspace(0, 1, num = 100)
# y <-- Represents original sine wave of 100 data points
y = np.sin(2*np.pi*xf)

# Plot

import matplotlib.pyplot as plot
plot.figure
plot.plot(x, t, 'bo',xf, y, 'g-')
plot.legend(('Noisy sampled 10 points', 'Sine wave'))
```

```python
# Fitting a polynomial curve on the blue points
# Input x and t (N = 10 Sampled points)
M = 3 # Order of the polynomial

z = np.polyfit(x, t, M) # Fitting the polynomial and finidng coefficients
p = np.poly1d(z) # Polynomial construction
y = p(x)

plot.figure
plot.plot(x, t, 'bo', x, y, 'r-')
plot.legend(('Training data', 'Model'))
```
